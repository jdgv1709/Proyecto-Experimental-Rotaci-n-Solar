```python
from scipy.constants import c
def vel_tan(real, doppler):
    #v= ((real/doppler)-1)*c
    v=((doppler-real)/real)*c
    return v
def tiempo(vt):
    vt_km= vt/1000 #metros a kilometros
    ra_sol= 1391016/2
    tiempo_s= (2*np.pi*ra_sol)/np.abs(vt_km)
    tiempo_d= tiempo_s/ (60*60*24)   #segundos a dias
    return tiempo_d
```
